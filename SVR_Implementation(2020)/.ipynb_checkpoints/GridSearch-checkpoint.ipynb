# GridSearch

```python
def grid_weekdays(X,y):
    #t_start = time.time()
    K = 5
    parameters = [{'kernel': ['rbf'], 
                   'gamma': ['auto','scale'],
                   'C': [0.1, 1, 10, 100, 1000, 10000], 
                   'epsilon': [0.05, 0.5, 1]
                  }]

    print("Tuning hyper-parameters...")
    scorer = make_scorer(mean_squared_error, greater_is_better=False)
    svr = svm.SVR()
    classifier = GridSearchCV(svr, parameters, cv = K, scoring=scorer)
    clf = classifier.fit(X.reshape(-1,1), y)
    #t_end = time.time()
    #search_time = t_end - t_start
    
    #if search_time > 60:
     #   search_time = int(search_time/60)
      #  print(search_time,'Minutes')
    #else:
     #   print(int(search_time),'Seconds')
    
    return clf
```

```python
classifier = []
for i in range(len(X)):
    classifier.append(grid_weekdays(X[i],y[i]))
```

```python
classifier[0]
```

```python
regr_best_param = []
for i in range(len(classifier)):
    regr_best_param.append(classifier[i].cv_results_['params'][classifier[i].best_index_])
    
regr_best_param[0]
```

```python
C = []
gamma = []
epsilon = []
for i in range(len(classifier)):
    C.append(regr_best_param[i].get('C'))
    gamma.append(regr_best_param[i].get('gamma'))
    epsilon.append(regr_best_param[i].get('epsilon'))
    
print(C)
print(gamma)
print(epsilon)
```
